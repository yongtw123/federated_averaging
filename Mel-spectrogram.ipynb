```python
import shelve
import librosa
import librosa.display
import matplotlib.pyplot as plt
```

```python
db = shelve.open('speech_commands_dataset_v2.cache' , 'r')
```

```python
zero_sample1 = db['zero/ffb86d3c_nohash_1.wav\n']
zero_sample2 = db['zero/0bd689d7_nohash_0.wav\n']
seven_sample = db['seven/0f46028a_nohash_0.wav\n']
```

```python
librosa.display.specshow(zero_sample1, sr=16000, x_axis='time', y_axis='mel');
plt.colorbar(format='%+2.0f dB');
```

```python
librosa.display.specshow(zero_sample2, sr=16000, x_axis='time', y_axis='mel');
plt.colorbar(format='%+2.0f dB');
```

```python
librosa.display.specshow(seven_sample, sr=16000, x_axis='time', y_axis='mel');
plt.colorbar(format='%+2.0f dB');
```
